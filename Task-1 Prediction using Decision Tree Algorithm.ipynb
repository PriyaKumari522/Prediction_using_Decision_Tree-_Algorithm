# Create the Decision Tree classifier and visualize it graphically

```python
# Importing libraries in Python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
#Reading the Dataset
data = pd.read_csv("../Downloads/Iris.csv")
```

```python
#check whether we have null values in our dataset or not.
data.isnull().any()
```

```python
#Displaying top rows of dataset 
data.head() #Deafault value of this function is 5
```

```python
#Divide the dataset into training and test dataset
all_inputs = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
all_classes = data['Species'].values

(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(all_inputs, all_classes, train_size=0.7, random_state=1)
```

We have to split the dataset into 70:30 for training and test dataset

```python
dtc = DecisionTreeClassifier()
dtc.fit(train_inputs, train_classes)
dtc.score(test_inputs, test_classes)
```

Let us visualize the Decision Tree to understand it better

```python
# Visualize the graph
dot_data = StringIO()
export_graphviz(dtc, out_file=dot_data, feature_names=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], 
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
